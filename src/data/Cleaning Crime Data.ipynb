```python
import folium
from folium.plugins import MarkerCluster
```

```python
crimeMap = folium.Map(
    location=[39.9509, -75.1875],
    zoom_start=13
)
```

```python
markClusters = MarkerCluster().add_to(crimeMap)
crimeMap
```

```python
locations = df[['lat', 'lng']]
locationList = locations.values.tolist()
for point in range(len(locationList)//250):
    popup = "<b>{}</b><br>{}<br>{}".format(df.location_block[point], \
                                         df.text_general_code[point], \
                                         df.dispatch_date_time[point])
    folium.Marker(locationList[point], popup=popup).add_to(markClusters)
    
#for i, j in df.iterrows():
#    popup = "<b>{}</b><br>{}<br>".format(j.location_block, j.text_general_code, j.dispatch_date_time)
#    folium.Marker([j.lat, j.lng], popup=popup).add_to(markClusters)
```
